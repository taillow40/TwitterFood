```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the CSV file
csv_path = 'training_loss.csv'
data = pd.read_csv(csv_path)

# Extract columns
epoch_train_loss = data['Loss']
epoch_val_loss = data['Val Loss']
epochs = data['Epoch']

# Plot the data
plt.plot(epochs, epoch_train_loss, label='Training Loss')
plt.plot(epochs, epoch_val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.ylim(0)
plt.legend(['Train', 'Val'], loc='upper left')
plt.title('Training and Validation Loss over Epochs')

# Save the plot

plt.savefig('Text_loss_word_embedding_plot.png')
plt.show()
```
